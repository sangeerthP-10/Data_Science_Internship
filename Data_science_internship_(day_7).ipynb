Break and Continue

```python
for i in range(10):
  if i==5:
    break
  print(i)
```

```python
i=1
while i<=10:
  print(f"6 * {i} = {6*i}" )
  if i>5:
    break
  i=i+1
```

```python
for i in range(10):
  if i==3 or i==6:
    continue
  print(i)
```

```python
#print even number
a=0
while a<10:
  a+=1
  if a%2!=0:
    continue
  print(a)
```

```python
#print odd number
a=0
while a<10:
  a+=1
  if a%2==0:
    continue
  print(a)
```

List

```python
fruits=["apple","mango","orange","banana"]
print(fruits)
print(len(fruits))
print(type(fruits))
```

```python
for fruit in fruits:
  print(fruit)
```

```python
print(fruits[0])
print(fruits[-1])
print(fruits[0:])
print(fruits[3:])
print(fruits[::-1])
print(fruits[::2])
print(fruits[-2:])
print(fruits[0:3])
```

Local and Global Variables

Global Variable -Declared outside code section,can be accesed throughout

Local Variable - Decared inside a function , can be accesed within the function

```python
f=100
print(f)

def func1():
  f=50
  print(f)

func1()


def func2():
  f=200
  print(f)

print(f)



func2()
```
